```python
from sklearn.ensemble import GradientBoostingRegressor

n = 5

rf = GradientBoostingRegressor(n_estimators=n
                            , learning_rate=0.5
                            , max_depth=8
                            , verbose=1
                            , warm_start=True)
```

#### Traing regression model in chunks

```python
import pandas as pd
from sklearn.model_selection import train_test_split

years = [2011, 2012, 2013, 2014, 2015]
#years = [2011]
x_test = pd.DataFrame()
y_test = pd.Series()

for y in years:
    print("fitting: " + str(y))
    df_chunk = pd.read_parquet('../../data/processed/features.parquet', engine='pyarrow', filters=[("year", "==", y)])
    target = df_chunk.pop('revenue')
    x_train, x_test_chunk, y_train, y_test_chunk = train_test_split(df_chunk, target, test_size=0.2, random_state=8)

    if x_test.empty:
        x_test = x_test_chunk
        y_test = y_test_chunk
    else:
        x_test = pd.concat([x_test, x_test_chunk])
        y_test = pd.concat([y_test, y_test_chunk])

    rf.fit(x_train, y_train)
    n = n + 5
    rf.set_params(n_estimators = n)

    print(rf.get_params())


```

```python
from joblib import dump
dump(rf, '../../models/predictive/xgb.joblib')
```

```python
y_pred = rf.predict(x_test)
```

```python
## RMSE evaluation
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test, y_pred, squared=False)
rmse
```

```python
# r2 score
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
