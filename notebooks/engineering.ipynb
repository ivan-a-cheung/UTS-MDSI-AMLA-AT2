```python
## use melt to unpvit the days
train_clean = pd.melt(train, id_vars=['item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'], var_name = 'd', value_name = 'items_sold')
```

```python
train_clean
```

```python
train_clean_2 = pd.merge(train_clean, df_cal, on='d', how='left')
#train_clean_2 = pd.merge(train_clean_2, df_events, on='date', how='left')
```

```python
train_clean_2.pop('d')
```

```python
train_clean_2 = pd.merge(train_clean_2, items, left_on = ['store_id', 'item_id', 'wm_yr_wk'], right_on = ['store_id', 'item_id', 'wm_yr_wk'], how = 'left')
```

```python
train_clean_2
```

```python
train_clean_2.pop('wm_yr_wk')
```

```python
train_clean_2['items_sold'] = train_clean_2['items_sold'].fillna(0)
train_clean_2['sell_price'] = train_clean_2['sell_price'].fillna(0)
```

```python
train_clean_2['revenue'] = train_clean_2['items_sold'] * train_clean_2['sell_price']
```
