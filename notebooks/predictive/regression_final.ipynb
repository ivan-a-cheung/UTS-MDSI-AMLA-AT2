### Encode categorical features

```python
df_clean
```

```python
cat_cols = df_clean.select_dtypes('category').columns
cat_cols
```

```python
df_clean = df_clean.loc[:, cat_cols]
df_clean.to_parquet('../../data/processed/df_clean_cat_cols.parquet')
```

```python
df_clean = pd.read_parquet('../../data/processed/df_clean_cat_cols.parquet')
```

```python
df_clean.info()
```

```python
df_clean
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1, dtype=int)
```

```python
oe.fit(df_clean[['item_id','dept_id', 'store_id']])
```

```python
df_clean[['item_id','dept_id', 'store_id']] = oe.transform(df_clean[['item_id','dept_id', 'store_id']])
```

```python
from joblib import dump
dump(oe, '../../models/item_encoder.joblib')
```

```python
cat_cols = df_clean.select_dtypes('category').columns
df_clean[cat_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, dtype=int, handle_unknown='infrequent_if_exist')
ohe.set_output(transform='pandas')

features = ohe.fit_transform(df_clean[cat_cols])
```

```python
dump(ohe, '../../models/ohe_encoder.joblib')
```

```python
features = features.join(df_clean[['item_id','dept_id', 'store_id']])
```

```python
features
```

```python
### recover memory
del(cat_cols)
del(df_clean)
del(oe)
del(ohe)
```

```python
df_clean_num = pd.read_parquet('../../data/processed/df_clean_num_cols.parquet')
```

```python
features = features.join(df_clean_num)
del(df_clean_num)
```

```python
features
```

```python
features.to_parquet('../../data/processed/features.parquet')
```

```python
from sklearn.model_selection import train_test_split
target = features.pop('revenue')
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=8)

del(features)
del(train_test_split)
```
